```python
# Top 100 free apps in Apple App Store, 24-01-20
url = "https://apps.apple.com/us/charts/iphone/top-free-apps/36"
```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import time
```

```python
page = requests.get(url)
doc_soup = BeautifulSoup(page.content, "html.parser")
```

```python
app_link = []
for app in doc_soup.find_all(class_="we-lockup--in-app-shelf"):
    app_link.append(app.a['href'])
len(app_link)
```

```python
app_dic = {}
app_url = app_link[3]
page = requests.get(app_url)
doc_soup = BeautifulSoup(page.content, "html.parser")
```

```python
app_dic['Name'] = doc_soup.find('h1', class_='product-header__title').text.strip().split('\n')[0]
app_dic['Age_rating'] = doc_soup.find('h1', class_='product-header__title').text.strip().split('\n')[1]
app_dic['Seller'] = doc_soup.find('h2', class_='app-header__identity').text.strip().split('\n')[0]
app_dic['Price'] = doc_soup.find('li', class_='app-header__list__item--price').text.strip().split('\n')[0]
app_dic['Category'] = doc_soup.find(class_='section--information').find('a', href=True).text.strip().split('\n')[0]
app_dic['Privacy_policy'] = doc_soup.find(class_='app-privacy').a['href']
```

```python
app_dic_list = []
for app_url in app_link:
    app_dic = {}
    page = requests.get(app_url)
    doc_soup = BeautifulSoup(page.content, "html.parser")
    app_dic['Name'] = doc_soup.find('h1', class_='product-header__title').text.strip().split('\n')[0]
    app_dic['Age_rating'] = doc_soup.find('h1', class_='product-header__title').text.strip().split('\n')[1].strip()
    app_dic['Seller'] = doc_soup.find('h2', class_='app-header__identity').text.strip().split('\n')[0]
    app_dic['Price'] = doc_soup.find('li', class_='app-header__list__item--price').text.strip().split('\n')[0]
    app_dic['Category'] = doc_soup.find(class_='section--information').find('a', href=True).text.strip().split('\n')[0]
    app_dic['Privacy_policy'] = doc_soup.find(class_='app-privacy').a['href']
    app_dic_list.append(app_dic)
app_dic_list[-5]
```

```python
df = pd.DataFrame(app_dic_list)
df.head()
```

```python
df.to_csv("Top100_Free_01-20.csv", index=False)
```
